```python
def split_latlong(result):
    if type(result) == float:
        return '0'
    else:
        result = result.replace('(','').replace(')','')
        result = result.split(',')
        api = request.get("https://geo.fcc.gov/api/census/area?lat=" + result[0] + "&lon=" + result[1] +"&format=json")
        county = api.json()['results'][0]['county_name']
        state = api.json()['results'][0]['state_code']

        return county + ', ' + state
```
